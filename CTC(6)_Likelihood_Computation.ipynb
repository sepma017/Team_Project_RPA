# Likelihood Computation
- 우리의 관심은 입력 음성 피처 시퀀스 x가 주어졌을 때 레이블 시퀀스 $l$이 나타날 확률, 즉 우도(likelihood)를 최대화하는 모델 파라메터(parameter)를 찾는 데 있다(Maximum Likelihood Estimation).
- 입력 음성 피처의 각 프레임에 대해서 모두 레이블이 있다면 지금까지 어렵사리 전방확률, 후방확률 구할 필요 없이 네트워크 마지막 레이어의 출력인 음소 확률 베터에서 정답 레이블에 관련된 확률들을 높여주면 된다(=MLE=Cross Entropy 최소화). 
- 하지만 우리가 가진 데이터는 입력 음성 시퀀스 x와 레이블 시퀀스 $l$ 뿐이다.
- 둘의 길이가 다르기 때문에 크로스 앤트로피를 적용할 수 없다. (그래서 먼 길을 돌아온 것이다.)

- 우도를 최대화 하려면 먼저 우도 값을 구해야 한다.
- CTC 문제에서 우도는 입력 데이터 시퀀스 x가 주어졌을 때 blank없는 원래 레이블 시퀀스 $l$이 나타날 확률을 가리킨다. (예시 기준으로 'h,e,l,l,o'가 나타날 확률)
- 이와 별개로 $l^\prime$은 blank가 포함된 레이블 시퀀스다.
- 우도, 'h,e,l,l,o'가 나타날 확률을 구하려면 'hel-lo--'에서부터 '--hel-lo'에 이르는 $l^\prime$에 관한 모든 가능한 경로들(그림11에서 회색 화살표의 모든 경우의 수)의 확률을 모두 고려(sum)해야 한다.

- 그런데 우도 계산 역시 전방확률과 후방확률이 있으면 편하게 구할 수 있다.
- 수식15 계산 겨로가는 blank가 포함된 레이블 시퀀스 $l^\prime$가운데 $t = 3$ 시점에 상태 $h$를 지나는 경로들에 대한 확률이다.
- 이 경로들은 전체 경로 가운데 일부에 해당한다(그림10 참조).
- $l^\prime$ 관련 전체 경로의 확률 합이 우도
- $t = 3$을 기준으로 우도를 구한다고 하면 그림11의 파란색으로 칠한 다섯 칸에 해당하는 Complete Path 확률을 모두 더한 값이 된다.(수식16과 같다.)

- 우도 계산을 일반화한 식은 수식17과 같다. 
- 시점(time)을 고정해 놓고 상태(state)에 대해 모두 합을 취하면 우도가 된다.
- 식에서 알 수 있듯이 어떤 시점에서 계산하든 우도 값은 동일하다.
- 각 시점, 상태일 확률이 조건부 독립(conditional independence)임을 가정하기 때문이다.
