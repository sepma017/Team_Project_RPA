# Gradient Computation
- 우도, 즉 $p(l|x)$를 최대화하는 모델 파라메터를 찾고자 한다.
- 이를 위해서는 우도에 대한 그레디언트(gradient)를 구해야 한다.
- 이 그래디언트를 모델 전체 학습 파라메터에 역전파(backpropagation)하는 것이 CTC 기법을 적용한 모델의 학습(train)이 된다.
- 우도 계산은 모통 로그우도(log-likelihood)로 수행한다.
- $t$번째 시점 $k$번째 상태(음소)에 대한 로그 우도의 그래디언트는 수식18과 같다.
- ln$(x)$를 $x$로 미분하면 $1/x$이고 우도에 로그를 취한 것을 합성 함수라고 이해하면 체인룰(chain)에 의해 수식18 우변처럼 정리할 수 있다.

- 수식18의 우변 마지막항 일부를 다시 정리한 결과는 수식19와 같다.
- 우도는 수식17과 같고 상수 $a$에 대해 $a/x$를 $x$로 미분한 결과는 $-a/x^2$라는 점을 고려하면 수식19 우변을 이해할 수 있다.
- 수식19에서 lab$(l,k)$라는 $k$라는 음소 레이블이 $l^\prime$에서 나타난 위치들을 가리킨다.

- 그림12에서 구하고자 하는 것은 $∂p(ln(1|x))/∂y^3_h$이다. (파란색이 칠해져 있는 칸에 해당하는 그래디언트)
- 이 값을 구하려면 우선 우도 $p(l|x)$부터 구해야 한다.
- 수식20과 같다.
