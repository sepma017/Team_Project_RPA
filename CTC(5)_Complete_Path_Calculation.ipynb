# Complete Path Calculation
- 특정 시점, 특정 상태를 지나는 경로에 대한 등장 확률을 구한다.
- 앞서 구한 전방확률과 후방확률로 간단히 계산할 수 있다.
- 그림10에서 파란색으로 칠한 칸을 본다.
- $t = 3$시점에 상태가 h$(s = 2)$인 모든 경로에 대한 확률을 구한다.
- 해당 경로는 '--hel-lo', '-hhel-lo', '-hhel-lo', 'hhhel-lo' 3가지가 존재한다.

프로젝트_이론_4.18.3.png

- '--hel-lo', '-hhel-lo', 'hhhel-lo' 3가지 경로에 대한 확률은 물론 모든 시점, 상태에 대해 일일이 곱셈을 하여서 구할 수 있다.
- 하지만 지금까지 설명했던 전방확률, 후방확률을 활용하면 좀 더 효율적으로 구할 수 있다.
- 파란색 칸에 대한 전방확률과 후방확률은 각각 수식12, 수식13과 같다.

프로젝트_이론_4.18.4.png

- 수식14는 수식12의 전방확률과 수식13의 후방확률을 곱한 것
- 식을 자세히 보면 전방확률과 후방확률의 곱은 우리가 구하고자 하는 '--hel-lo', '-hhel-lo', 'hhhel-lo' 3가지 경로에 대한 확률을 더한 값에 $y^3_h$이 곱해진 형태임을 확인할 수 있다.
- 수식11의 양변을 $y^3_h$으로 나눠주면 우리가 구하고자 하는 값을 구할 수 있다.
